**receieve location from name**

```python
name = pd.DataFrame({"name":['หางดง', 'ยางเนิ้ง', 'สันกำแพง','สันทราย', 'แยกศาลเด็ก'],
                     "value":[20, 38, 40, 30, 10]}, dtype=str)
l =[]
for i in range(0,len(name)):
    location = name.iloc[i]['name']
    loc = geocoder.osm(location)
    radius = name.iloc[i]['value']
    latlng = [loc.lat, loc.lng, location, radius]
    l.append(latlng)
location_latlng = pd.DataFrame(l, columns=["latitude", "longitude","name", "radius"])
location_latlng
```

**create radius from location**

```python
m = folium.Map(location=[18.7883, 98.9853], tiles="OpenStreetMap", zoom_start=11)
for i in range(0,len(location_latlng)):
    folium.CircleMarker(
        location=[location_latlng.iloc[i]['latitude'], location_latlng.iloc[i]['longitude']],
        radius = [location_latlng.iloc[i]['radius']],
        color = '#FE0000',
        fill = True,
        fill_color = '#E3242B',    
   ).add_to(m)
m
```
