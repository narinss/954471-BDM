```python
#ex4
rng = np.random.default_rng()
data_matrix = rng.integers(10, size=(10,5))
label_vector = np.random.randint(2,size = 10)
train_data = data_matrix[np.where(label_vector == 0)]
test_data = data_matrix[np.where(label_vector == 1)]
print ("data_matrix" + "\n" + str(data_matrix) + "\n" + "\n" 
       + "label_vector" + "\n" + str(label_vector) + "\n" + "\n"
       + "train_data" + "\n" + str(train_data) + "\n" + "\n"
       + "test_data" + "\n" + str(test_data) + "\n")
```
